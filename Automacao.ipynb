## Atomação de Tarefas e Processos com Python

Neste notebook vamos desenvolver um código afim de atomatizar o processo de envio de um relatório com todas as compras de mercadorias de uma empresa fictícia. 

O trabalho será enviar um email com total de gasto, a quantidade de produtos compradas e o preço médio dos produtos.

Usaremos o pyautogui, uma biblioteca de automação de comandos do mouse e teclado para resolver esta tarefa

```python
# !pip install pyautogui
```

```python
import pyautogui
import time
import pandas as pd
import pyperclip

pyautogui.PAUSE = 1

# Acessando a base de dados

pyautogui.click(x=626, y=24)  # pyautogui.position()
pyautogui.write("https://drive.google.com/drive/folders/1iUbm54CKTYrSpuVo-Dgu7zrWObsKMM1t")
pyautogui.press("enter")

time.sleep(2)

# Baixando a base de  dados

pyautogui.click(x=383, y=411)
pyautogui.click(x=634, y=202)

time.sleep(2)

# Importando a base de dados

tabela = pd.read_csv(r"C:\Users\Isabella Nunes\Downloads\Compras.csv", sep=";")

# Calculando indicadores

total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade

# Enviando email com os indicadores

pyautogui.click(x=935, y=14)
pyautogui.write("gmail.com")
pyautogui.press("enter")

time.sleep(2)

pyautogui.click(x=107, y=201)
pyautogui.write("isabellarsnunes@gmail.com")
pyautogui.press("tab")
pyautogui.press("tab")
pyautogui.write("Dados de Vendas")
pyautogui.press("tab")

texto = f"""
Prezados(as),

Segue relatório com dados de compra.

Total Gasto: R$ {total_gasto:,.2f}
Quantidade: {quantidade:,}
Preço Médio: R$ {preco_medio:,.2f}

Para mais informações, entre em contato.

Att,
Isabella Nunes 
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

time.sleep(2)

# Enviar

pyautogui.hotkey("ctrl", "enter")
```
